# Ejercicio 4

```julia
A=[-1 1; -1 -1];
```

```julia
t=symbols("t");
```

```julia
f(t)=expm(A*t);
```

```julia
x=[1;2]
listx=Float64[1]
listy=Float64[2]
for i in linspace(0,10,1000)
    d=f(i)*x
    q=d[1]
    w=d[2]
    push!(listx,q)
    push!(listy,w)
end
plot(listx,listy);
```
