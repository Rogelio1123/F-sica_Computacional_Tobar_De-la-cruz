```julia
function derivada_numerica(f,x,h)
    g=(f(x+h)-f(x))/h
    return g
end;
```

# Ejercicio 2

```julia
derivada_numerica(cosd,0.0001,0.01)
```

El resultado exacto de la derivada anterior es -1.7453292519934436e-6

A continuación se grafica el valor de la derivada númerica respecto a $1/h$

```julia
y=linspace(0.001,1,1000);
z=1/y;
```

```julia
L=zeros(1000)
for i in 1:1000
    g=y[i]
    L[i]=derivada_numerica(cosd,0.0001,g)
end 
```

```julia
xlabel("1/h")
ylabel("derivada númerica")
title("Derivada númerica Vs inverso de h")
grid("on")
plot(z,L);
```

A continuación se grafica el error relativo definido como $\epsilon=\mid \frac{\frac{ f(x_0 +h)-f(x_0)}{h} -f'(x_0)}{f'(x_0)} \mid$ respecto a h.

```julia
E=zeros(1000)
for i in 1:1000
    g=y[i]
    E[i]=abs((((cosd(0.0001+g)-cosd(0.0001))/g)+sind(0.0001))/(-sind(0.0001)))
end 
```

```julia
xlabel("h")
ylabel("Error relativo")
title("Error relativo Vs h (derivada_numerica)")
grid("on")
plot(y,E);
```

En la grafica anterior se puede ver que el error relativo es del orden de h.

# Ejercicio 3

Puesto que:
\begin{equation}
f(x+h)=f(x)+h\frac{d}{ds} f(s) \mid_{s=x} + \frac{h^2}{2} \frac{d^2}{ds^2} f(s) \mid _{s=x} + O(h^3)
\end{equation}
Se tiene que:
\begin{equation}
f(x-h)=f(x)-h\frac{d}{ds} f(s) \mid_{s=x} + \frac{h^2}{2} \frac{d^2}{ds^2} f(s) \mid _{s=x} + O(h^3)
\end{equation}
Entonces:
\begin{equation}
f(x+h)-f(x-h)=2h\frac{d}{ds} f(s) \mid_{s=x} + O(h^3)
\end{equation}
Por lo tanto: 
\begin{equation}
\frac{f(x+h)-f(x-h)}{2h} - \frac{d}{ds} f(s) \mid_{s=x}= O(h^2)
\end{equation}
Como en este caso el error es del orden de $h^2$ esta nueva forma de calcular la derivada mejora a la derivada hacía adelante.

# Ejercicio 5

Se define $\Delta_h :f \to f(x+h)-f(x)$, 
entonces se tiene lo siguiente:

\begin{equation}
\Delta_h (\alpha f+g)=(\alpha f+g)(x+h)-(\alpha f+g)(x)=\alpha f(x+h) + g(x+h) -\alpha f(x)-g(x)= \alpha \Delta_h f +\Delta g 
\end{equation}

Por lo tanto $\Delta_h$ es lineal.

Por otro lado, se tiene la siguiente relación:

\begin{equation}
\Delta_h f = f(x+h)-f(x) = f(x)+h\frac{d}{ds} f(s) \mid_{s=x} + \frac{h^2}{2} \frac{d^2}{ds^2} f(s) \mid _{s=x} + O(h^3) -f(x)=h\frac{d}{ds} f(s) \mid_{s=x} + \frac{h^2}{2} \frac{d^2}{ds^2} f(s) \mid _{s=x} + O(h^3)
\end{equation}

Lo que implica: 

\begin{equation}
\Delta_h f = \sum_{n=1}^\infty \frac{1}{n!} h^n \frac{d^n}{ds^n} f(s)  \mid _ {s=x}
\end{equation}
De donde se obtiene que:
\begin{equation}
\Delta_h  = \sum_{n=1}^\infty \frac{1}{n!} h^n \frac{d^n}{ds^n}  \mid_{s=x}
\end{equation}

# Ejercicio 6

Se considera lo siguiente:

\begin{equation}
e^{h\frac{d}{ds} \mid_{s=x}} = \sum_{n=0}^infty \frac{1}{n!}h^n\frac{d^n}{ds^n} \mid_{s=x} =1+\sum_{n=1}^infty \frac{1}{n!}h^n\frac{d^n}{ds^n} \mid_{s=x} =1+\Delta_h
\end{equation}

Por lo tanto:
\begin{equation}
\Delta_h = e^{h\frac{d}{ds} \mid_{s=x} } -1
\end{equation}
