# Ejercicio 2

La ecuación de laplace en una dimensión en diferencias finitas queda de la siguiente forma:
\begin{equation}
\frac{f_{i+1}+f_{i-1}-2f_i}{h^2}=\rho_i
\end{equation}
Con el dominio $[0,2\pi]$, usando 12 puntos y $\rho(x)=cos(x)$ se obtiene el sistema
\begin{eqnarray}
f_2-2f_1=(\frac{\pi}{6})^2cos(1\frac{2\pi}{12})\\
f_3+f_1-2f_2=(\frac{\pi}{6})^2cos(2\frac{2\pi}{12})\\
f_4+f_2-2f_3=(\frac{\pi}{6})^2cos(3\frac{2\pi}{12})\\
f_5+f_3-2f_4=(\frac{\pi}{6})^2cos(4\frac{2\pi}{12})\\
f_6+f_4-2f_5=(\frac{\pi}{6})^2cos(5\frac{2\pi}{12})\\
f_7+f_5-2f_6=(\frac{\pi}{6})^2cos(6\frac{2\pi}{12})\\
f_8+f_6-2f_7=(\frac{\pi}{6})^2cos(7\frac{2\pi}{12})\\
f_9+f_7-2f_8=(\frac{\pi}{6})^2cos(8\frac{2\pi}{12})\\
f_{10}+f_8-2f_9=(\frac{\pi}{6})^2cos(9\frac{2\pi}{12})\\
f_{11}+f_9-2f_{10}=(\frac{\pi}{6})^2cos(10\frac{2\pi}{12})\\
f_{12}+f_{10}-2f_{11}=(\frac{\pi}{6})^2cos(11\frac{2\pi}{12})\\
f_{11}-2f_{12}=(\frac{\pi}{6})^2cos(12\frac{2\pi}{12})\\
\end{eqnarray}
que se convierte en:
\begin{eqnarray}
f_2-2f_1=0.2374\\
f_3+f_1-2f_2=0.1370\\
f_4+f_2-2f_3=0\\
f_5+f_3-2f_4=-0.1370\\
f_6+f_4-2f_5=-0.2374\\
f_7+f_5-2f_6=-0.2741\\
f_8+f_6-2f_7=-0.2374\\
f_9+f_7-2f_8=-0.1370\\
f_{10}+f_8-2f_9=0\\
f_{11}+f_9-2f_{10}=0.1370\\
f_{12}+f_{10}-2f_{11}=0.2374\\
f_{11}-2f_{12}=0.2741\\
\end{eqnarray}
