```julia
function Tomografia_qubit(D,n,A)
    x=[1;0]
    y=[0;1]
    if A == 3
        P=x*x'
    elseif A == 1
        k=(1/sqrt(2))*(x+y)
        P=k*k'
    elseif A==2 
        k=(1/sqrt(2))*(x+1im*y)
        P=k*k'
    end
    h=P*D
    q=0
    for i in 1:2
        q=q+h[i,i]
    end
    list=zeros(n)
    for i in 1:n
    l=rand()
    if l<=abs(q)
        list[i]=1
    else
            list[i]=-1
    end
    end
    return list
end;
```

```julia
D=[1/2 1/2; 1/2 1/2];
```

```julia
using PyPlot;
```

```julia
s=Tomografia_qubit(D,100,1)
nbins = 2 
fig = figure("pyplot_histogram",figsize=(1,5)) 
ax = axes() 
h = plt[:hist](s,nbins) 
grid("on")
xlabel("Estado")
ylabel("Numero de eventos")
title("Histogram");
```
