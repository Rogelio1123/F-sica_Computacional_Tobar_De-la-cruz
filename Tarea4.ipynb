```julia
using PyPlot;
```

```julia
function convergencia(A)
    list=zeros(20)
    x=A
    for i in 1:20
        f=x^3-x^2-x-1
        df=3x^2-2x-1
        x=x-f/df
        list[i]=x
    end 
    return list
end;
```

```julia
y=convergencia(4)
z=convergencia(5)
w=convergencia(8)
q=convergencia(11);
```

A continuación se grafica la convergencia de 4 condiciones iniciales usando el método de Newton. 

```julia
yscale("symlog", linthreshy=11)
plot(real(y))
plot(real(z))
plot(real(w))
plot(real(q))
show();
```

Ahora se grafica el error relativo a la raiz en cada iteración.

```julia
yscale("symlog", linthreshy=1)
plot(real((y-1.8392867552141612)/1.8392867552141612))
plot(real((z-1.8392867552141612)/1.8392867552141612))
plot(real((w-1.8392867552141612)/1.8392867552141612))
plot(real((q-1.8392867552141612)/1.8392867552141612))
show();
```

```julia
function biseccion(a,b)
    p = (a+b)/2
    f(x) = x^3-x^2-x-1
    list=zeros(50)
    for i in 1:50
        if  f(a)*f(p) > 0
            a = p
        end
        
        if f(b)*f(p) > 0
            b = p
        end
        p = (a+b)/2
        list[i]=p
    end
    return list
end;
```

```julia
t=biseccion(0,5)
h=biseccion(0,20)
b=biseccion(0,12)
n=biseccion(0,7);
```

Se grafica la convergencia de 4 condiciones iniciales usando el método de bisección.

```julia
yscale("symlog", linthreshy=11)
plot(real(t))
plot(real(h))
plot(real(b))
plot(real(n))
show();
```
