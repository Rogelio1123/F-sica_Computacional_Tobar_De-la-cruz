# Ejercicio 1

Se genera una rutina que implemente el método de euler

```julia
function metodo_euler(f,x0,t0,tf,h)
    listt=Float64[t0]
    listx=Float64[x0]
    t=t0
    i=1
    while i<(tf-t0)/h
        x=listx[i]
        push!(listx,x+(f(x,t))*h)
        t=t+h
        push!(listt,t)
        i+=1
    end
    return listt,listx
end;
```

#### Test:

Se resolverá la ecuación $\frac{dx}{dt} = - \alpha x$ empleando las siguientes condiciones: $\alpha=-2.5$, $x_0={1,2}$, $y \epsilon (0,8)$ y $h={1.0,0.5,0.2,0.01}$

```julia
f(x,t)=-2.5*x;
```

```julia
using PyPlot
```

```julia
w1,q1=metodo_euler(f,1,0,8,1);
```

```julia
xlabel("t")
ylabel("x")
title("Comparacion entre la solución exacta para x(0)=1 y la aproximación con h=1")
grid("on")
plot(w1,exp.(-2.5*w1),color="blue")
plot(w1,q1,color="red")
show();
```

# Ejercicio 2

El oscilador armónico es modelado por la ecuación $\ddot{x}=-kx$, la solución general a dicha ecuación tiene la forma:
\begin{equation}
x(t)=Asen(\sqrt{k}t)+Bcos(\sqrt{k}t)
\end{equation}
Si se consideran las siguiente condiciones iniciales $x(0)=x_0$ y $\dot{x}(0)=\dot{x}_0$, se tiene lo siguiente:
\begin{equation}
x_0=B  \quad y \quad A\sqrt{k}=\dot{x}_0
\end{equation}
De donde se concluye:
\begin{equation}
x(t)=\frac{\dot{x}_0}{\sqrt{k}}sen(\sqrt{k}t)+x_0cos(\sqrt{k}t)
\end{equation}
